```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# membaca gambar
img = cv2.imread('input.jpeg', 0)

# cek apakah gambar terbaca
if img is None:
    print("Gambar tidak ditemukan!")
else:

    # Negative
    negative = 255 - img

    # Log Transform
    img_float = img.astype(np.float32)
    c = 255 / np.log(1 + np.max(img_float))
    log_transform = c * np.log(1 + img_float)
    log_transform = np.array(log_transform, dtype=np.uint8)

    # Power Law
    gamma = 2
    power_law = np.array(255*(img/255)**gamma, dtype='uint8')

    # tampilkan hasil
    plt.figure(figsize=(10,8))

    plt.subplot(2,2,1)
    plt.imshow(img, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(2,2,2)
    plt.imshow(negative, cmap='gray')
    plt.title("Negative Image")
    plt.axis("off")

    plt.subplot(2,2,3)
    plt.imshow(log_transform, cmap='gray')
    plt.title("Log Transform Image")
    plt.axis("off")

    plt.subplot(2,2,4)
    plt.imshow(power_law, cmap='gray')
    plt.title("Power Law Image")
    plt.axis("off")

    plt.show()
```
